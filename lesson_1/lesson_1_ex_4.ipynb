```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_sin(k, x):
    return np.sin(k*x)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 201)

koef = (1, 0.5)

for k in koef:
  plt.plot(x, func_sin(k, x) ) 

plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.show()
```
