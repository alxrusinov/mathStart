```
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```
def get_vector_length(start, end):
  result = 0
  for start_coord, end_coord in zip(start, end):
    result += (end_coord - start_coord)**2
  return math.sqrt(result)
```

```
vector_1 = ((0,0,0), (5,5,5))
vector_2 = ((2,3,4),(8,9,9))
print(get_vector_length(*vector_1))
print(get_vector_length(*vector_2))
```

## Задание 3

```
# полярные в декартовы
def polar_to_decart(coord):
  R, angle = coord
  return R * math.cos(angle), R * math.sin(angle)

print(polar_to_decart((1,np.pi/4)))  
  
```

```
# окружность в полярных координатах
def circle_in_polar(r):
  x = np.linspace(0, np.pi * 2, 100)
  y = np.full(x.shape, r)
  plt.polar(x, y)

circle_in_polar(23)
```

```
# отрезок прямой в полярных координатах
def direct_in_polar(coord):
  p, angle = coord
  x = np.linspace(0, np.pi / 2, 500)
  y = p / np.cos(x - angle)
  plt.polar(x, y)

direct_in_polar((2, np.pi/4))
  
```

## Задание 4

```
# y = x**2 – 1 => y - x**2 + 1 = 0 => y = x**2 – 1
# exp(x) + x*(1 – y) = 1 => exp(x) + x*(1 - y) - 1 = 0 => y = 1 - (1 - exp(x))/x
x = np.linspace(-15, 20, 1000)
plt.plot(x, x**2 - 1)
plt.plot(x, 1 - (1 - np.exp(x))/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 =  fsolve(equations, (-5, 1))
x2, y2 =  fsolve(equations, (0.1, 2))
              
print (x1, y1)
print (x2, y2)
```

```
# y = x**2 – 1 => y - x**2 + 1 => y = x**2 – 1
# exp(x) + x*(1 – y) > 1 => exp(x) + x*(1 - y) - 1 > 0 => y < 1 - (1 - exp(x)) / x
x = np.linspace(-15, 20, 1000)
plt.plot(x, x**2 - 1)
plt.plot(x, 1 - (1 - np.exp(x))/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def condition(x, y):
  return y < 1 - (1 - math.exp(x)) / x

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

def get_result(x, y):
    x1, y1 =  fsolve(equations, (x, y))
    if condition(x1, y1):
      return x1, y1
    else:
      return None, None  

x1, y1 =  get_result(-5, 1)
x2, y2 =  get_result(0.1, 2)
              
print (x1, y1)
print (x2, y2)
```
